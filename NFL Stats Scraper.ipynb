```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import pymysql

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")

# For use, change the Year variable to suit your needs.

Data_Year = "2023"
service = Service(ChromeDriverManager().install())
```

```python
driver = webdriver.Chrome(service=service, options=chrome_options)
driver.get('https://www.pro-football-reference.com/years/'+Data_Year+'/passing.htm')


ranks=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/th[@scope="row"]')
players=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[1]/a')
teams=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[@data-stat="team"]')
ages=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[3]')
positions=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[4]')
gamesplayed=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[5]')
gamesstarted=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[6]')
passescompleted=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[8]')
passesattempted=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[9]')
passescompleteperc=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[10]')
pass_yards=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[11]')
pass_td=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[12]')
pass_td_perc=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[13]')
pass_int=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[14]')
pass_int_perc=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[15]')
pass_first_down=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[16]')
pass_long=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[17]')
pass_yds_per_attempt=driver.find_elements("xpath", '//*[@id="passing"]/tbody/tr/td[18]')

Passing_stats=[]

print(len(players))

for i in range(len(players)):
    Pass_temp_data={'Rank': ranks[i].text,
                    'Player': players[i].text,
                   'Team': teams[i].text,
                   'Age': ages[i].text,
                   'Position': positions[i].text,
                   'Games Played': gamesplayed[i].text,
                   'Games Started': gamesstarted[i].text,
                   'Passes Attempted': passesattempted[i].text,
                   'Passes Completed': passescompleted[i].text,
                   'Passes Completed Percentage': passescompleteperc[i].text,
                   'Pass Yards': pass_yards[i].text,
                   'Passing Touchdowns': pass_td[i].text,
                   'Passing Touchdown Percentage': pass_td_perc[i].text,
                   'Passing Interceptions': pass_int[i].text,
                   'Passing Interception %': pass_int_perc[i].text,
                   'Passing First Downs': pass_first_down[i].text,
                   'Longest Pass': pass_long[i].text,
                   'Passing Yards Per Attempt': pass_yds_per_attempt[i].text
               
                   
              }
    Passing_stats.append(Pass_temp_data)
df_player_data=pd.DataFrame(Passing_stats)
df_player_data['Year'] = Data_Year

#df_player_data.to_excel('NFL '+Data_Year+' passing Stats.xlsx',index=False)



driver.close();


```
